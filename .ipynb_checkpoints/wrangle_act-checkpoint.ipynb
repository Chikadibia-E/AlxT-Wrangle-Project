## Data Gathering
In the cell below, gather **all** three pieces of data for this project and load them in the notebook. **Note:** the methods required to gather each data are different.
1. Directly download the WeRateDogs Twitter archive data (twitter_archive_enhanced.csv)

```python
# Import python libraries to work with
import pandas as pd
import os
import requests
import tweepy
import configparser
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from io import BytesIO
from PIL import Image

%matplotlib inline
```

```python
# Read the given/downloaded twitter archive
twitter_archive = pd.read_csv('twitter-archive-enhanced.csv')
```

2. Use the Requests library to download the tweet image prediction (image_predictions.tsv)

```python
# create folder if it does ot exist
folder = 'image-predictions'
if not os.path.exists(folder):
    os.makedirs(folder)
    
# Download the image prediction file with request library
url ='https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
response
```

```python
# write the content of image prediction file to folder
with open (os.path.join(folder, url.split('/')[-1]), mode = 'wb') as f:
           f.write(response.content)
```

3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
# Read app/consumer key configuration with configparser library
config = configparser.ConfigParser()
config.read('config.ini')

api_key = config['twitter']['api_key']
api_key_secret = config['twitter']['api_key_secret']

access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']

# Authentication with twitter app keys 
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
# Getting required tweet contents with twitter API
with open('tweet_json.txt', 'w') as file:
    for i in twitter_archive['tweet_id']:
        try:
            tweets = api.get_status(i, tweet_mode='extended')
            json.dump(tweets._json, file)
            file.write('\n')
        except Exception as e:
            pass
```

```python
# # Read tweet_json text file line by line from json data
selected_cols = []
with open('tweet_json.txt', 'r') as json_file:
    for line in json_file:
        json_data = json.loads(line)
        
        # create a dictionary from json data, and form selected columns (tweet_id, favorites, retweets)
        selected_cols.append({'tweet_id': json_data['id'],
                             'favorites': json_data['favorite_count'],
                             'retweets': json_data['retweet_count']})
# convert the tweet JSON data dictionary list to a DataFrame
selected_tweet_data = pd.DataFrame(selected_cols, columns=['tweet_id','favorites','retweets'])
```

```python
# view selected columns
selected_tweet_data
```

### Tidy data issue 4

### Define

+ Multiple prediction columns, p1,p1_dog,p1_conf p2,p2_dog,p2_conf,p3,p3_dog,p3_conf. Dog predictions can be in one column

### Code

```python
# compact predictions to one column
def dog_prediction (predict):
    if predict['p1_dog']:
        return predict ['p1']
    elif predict['p2_dog']:
        return predict ['p2']
    elif predict['p3_dog']:
        return predict ['p3']
    else:
        return np.nan


df_merged['dog_prediction'] = df_merged.apply(dog_prediction, axis = 1)
```

```python
# check the just constructed dog_prediction column
df_merged.dog_prediction
```

```python
#drop columns not needed for analysis. drop rating_denominator too since the contents are all 10 
df_merged.drop(columns = ['p1','p1_dog','p1_conf', 'p2','p2_dog','p2_conf',
                          'p3','p3_dog','p3_conf','rating_denominator',
                         'img_num'], inplace = True)
```
